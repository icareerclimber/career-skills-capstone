```python
from gensim import corpora
import nltk
from nltk.corpus import stopwords

titleStopWords = set(stopwords.words('english'))
titleStopWords
```

```python
titles = [t.lower().split(' ') for t in uniqueTitles if type(t) is str]

def titleTransformSimple(t):
    if t == "sr" or t == "sr.":
        return 'senior'
    elif t == 'ii':
        return '2'
    elif t == '&':
        return 'and'
    elif len(t) > 0 and t[0] == '(':
        return t[1:]
    elif len(t) > 0 and t[-1] == ')':
        return t[:-1]
    else:
        return t
```

```python
titlesClean = map(lambda title: list(map(lambda word: titleTransformSimple(word), title)), titles)
titlesClean = [[word for word in title if word not in titleStopWords and len(word) > 0] for title in titlesClean]
list(titlesClean)
```

```python
from collections import defaultdict
frequency = defaultdict(int)
for title in titlesClean:
    for token in title:
        frequency[token] += 1
        
#TODO: handle / and (). Maybe duplicate the entry as two entries.

titlesClean = [[token for token in title if frequency[token] > 1] for title in titlesClean]
titlesClean
```

```python
titleDict = corpora.Dictionary(titlesClean)
```

```python
titleDict.save("../resumes/titleDictionaryAtlanta.dict")
```

```python
import gensim
word2vecModel = gensim.models.Word2Vec(titlesClean, min_count=1)
```

```python
word2vecModel.save("../resumes/word2vecAtlantaMinCount1Size100")
```

```python
def genVector(title):
    result = word2vecModel['engineer'] # default unknown job title to engineer
    try:
        result = np.sum(word2vecModel[title], axis=0)
    except:
        pass
    return result

titleVectors = [genVector(title) for title in titlesClean]
titleVectors
```

```python
from sklearn.cluster import KMeans

atlantaKMean = KMeans(n_clusters=5, random_state=0).fit(titleVectors)
```

```python
atlantaLabels = atlantaKMean.labels_
#print(atlantaKMean.cluster_centers_)
l0 = len([l for l in atlantaLabels if l == 0])
l1 = len([l for l in atlantaLabels if l == 1])
l2 = len([l for l in atlantaLabels if l == 2])
l3 = len([l for l in atlantaLabels if l == 3])
l4 = len([l for l in atlantaLabels if l == 4])

print("{} {} {} {} {}".format(l0, l1, l2, l3, l4))
```
