# Glassdoor Salaries

```python
datastore = []
directory = '/Users/kwheatley/Downloads/tech-salary-project/salaries/'
for path, subdirs, file_list in os.walk(directory):  
    for file in file_list:
        if file != '.DS_Store':
            try:
#                 print("file:", file)
                json_data=open(path + '/' + file).read()
                records_list = json.loads(json_data)
                for record in records_list:
                    meanPay = record['meanPay']
                    jobTitle = record['jobTitle']
                    company = record['company']
                    row = [meanPay,jobTitle,company]                
                    
                    with open("glassdoor_salaries.csv", 'a') as myfile:
                        wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
                        wr.writerow(row)
            except Exception as e:
                print(path + '/' + file)
                print(traceback.format_exc())
                continue
```
